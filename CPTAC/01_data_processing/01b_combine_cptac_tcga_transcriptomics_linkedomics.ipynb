```python
combined_transcriptomics = pd.concat([ccrcc_transcriptomics, endo_transcriptomics, luad_transcriptomics, 
                                      brca_2020_transcriptomics, gbm_transcriptomics, 
                                      hnscc_transcriptomics, ovca_2020_transcriptomics, 
                                      lscc_transcriptomics, pdac_transcriptomics, 
                                      brca_2016_transcriptomics, ovca_2016_transcriptomics], axis=1)

print("Dimensions: ", combined_transcriptomics.shape)
combined_transcriptomics[:2]
```

#### Filtering the proteins and samples

```python
# Restrict the data to contain only the protein-coding genes 
protein_coding_genes = pd.read_csv(file_protein_coding_genes, sep='\t')
protein_coding_genes = protein_coding_genes[(protein_coding_genes.status == 'Approved') & 
                                            (protein_coding_genes['Locus type'] == 'Gene with protein product')]['Symbol']
protein_coding_genes[:3]
```

```python
cptac_samples = pd.read_parquet(file_cptac_samples_info)
cptac_samples[:2]
```

```python
common_protein_coding_genes = np.intersect1d(combined_transcriptomics.index, list(protein_coding_genes))
common_samples = np.intersect1d(combined_transcriptomics.columns, cptac_samples.index)
print("Common samples = {0} and common proteins = {1}".format(len(common_samples), len(common_protein_coding_genes)))
combined_transcriptomics = combined_transcriptomics.reindex(common_protein_coding_genes).reindex(common_samples, axis=1)
print("Dimensions of combined CPTAC proteomics database: ", combined_transcriptomics.shape)
combined_transcriptomics[:5]
```

```python
combined_transcriptomics.to_parquet(get_tables_path('CPTAC_transcriptomics_linkedomics.parquet'))
```
